```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

data = np.load("ising_dataset.npz")
spin_configs = data["spin_configs"]        
psi_squared_true = data["psi_squared"]     

psi_squared_true /= psi_squared_true.sum()

X = torch.tensor(spin_configs, dtype=torch.float32)
y_true = torch.tensor(psi_squared_true, dtype=torch.float32).unsqueeze(1) 
```

```python
class PINN(nn.Module):
    def __init__(self, input_dim):
        super(PINN, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 128),
            nn.ReLU(),
            nn.Linear(128, 1),
            nn.Softplus() 
        )

    def forward(self, x):
        out = self.net(x)
        return out
```

```python
def ising_energy(spins, J=1.0):
    return -J * torch.sum(spins[:, :-1] * spins[:, 1:], dim=1, keepdim=True)
```

```python
def physics_loss(prob_pred, energies, true_ground_energy):
    exp_energy = torch.sum(prob_pred * energies)
    return (exp_energy - true_ground_energy) ** 2
```

```python
from scipy import sparse
from scipy.sparse.linalg import eigsh

def get_ground_state_energy(n_spins, J=1.0, h=1.0):
    dim = 2 ** n_spins
    H = sparse.lil_matrix((dim, dim))

    for idx in range(dim):
        config = [int(x) for x in format(idx, f'0{n_spins}b')]
        spin_config = [1 if s == 0 else -1 for s in config]
        
        diag_term = 0
        for i in range(n_spins - 1):
            diag_term -= J * spin_config[i] * spin_config[i + 1]
        H[idx, idx] = diag_term

        for i in range(n_spins):
            new_config = config.copy()
            new_config[i] = 1 - new_config[i]
            new_idx = int(''.join(map(str, new_config)), 2)
            H[idx, new_idx] = -h

    H_csr = H.tocsr()
    eigenvalues, _ = eigsh(H_csr, k=1, which='SA')
    return float(eigenvalues[0])

E0 = get_ground_state_energy(n_spins=spin_configs.shape[1], J=1.0, h=1.0)
E0 = torch.tensor(E0, dtype=torch.float32)
```

```python
# model = PINN(input_dim=X.shape[1])
# optimizer = optim.Adam(model.parameters(), lr=1e-3)
# epochs = 1000
# lambda_phys = 0.5  # Weight for physics-informed loss


# for epoch in range(epochs):
#     model.train()
#     optimizer.zero_grad()
    
#     # Forward
#     y_pred = model(X)
#     y_pred_norm = y_pred / torch.sum(y_pred)

#     # Supervised loss
#     mse_loss = nn.MSELoss()(y_pred_norm, y_true)

#     # Energy terms
#     energies = ising_energy(X)
#     phys_loss = physics_loss(y_pred_norm, energies, E0)

#     # Total loss
#     total_loss = lambda_phys * phys_loss + (1 - lambda_phys) * mse_loss
#     loss_log = {
#     "total": [],
#     "mse": [],
#     "phys": []
# }

#     loss_log["total"].append(total_loss.item())
#     loss_log["mse"].append(mse_loss.item())
#     loss_log["phys"].append(phys_loss.item())
#     total_loss.backward()
#     optimizer.step()

#     if epoch % 100 == 0:
#         print(f"Epoch {epoch}: Total Loss = {total_loss.item():.6f}, MSE = {mse_loss.item():.6f}, Physics = {phys_loss.item():.6f}")
# print(loss_log)

# model = PINN(input_dim=X.shape[1])
# optimizer = optim.Adam(model.parameters(), lr=1e-3)
# epochs = 1000
# lambda_phys = 0.00001  # Weight for physics-informed loss

# # Move this outside the training loop
# loss_log = {
#     "total": [],
#     "mse": [],
#     "phys": []
# }

# for epoch in range(epochs):
#     model.train()
#     optimizer.zero_grad()
    
#     # Forward
#     y_pred = model(X)
#     y_pred_norm = y_pred / torch.sum(y_pred)

#     # Supervised loss
#     mse_loss = nn.MSELoss()(y_pred_norm, y_true)

#     # Energy terms
#     energies = ising_energy(X)
#     phys_loss = physics_loss(y_pred_norm, energies, E0)

#     # Total loss
#     total_loss = lambda_phys * phys_loss + (1 - lambda_phys) * mse_loss

#     # Log losses (append)
#     loss_log["total"].append(total_loss.item())
#     loss_log["mse"].append(mse_loss.item())
#     loss_log["phys"].append(phys_loss.item())

#     total_loss.backward()
#     optimizer.step()

#     if epoch % 100 == 0:
#         print(f"Epoch {epoch}: Total Loss = {total_loss.item():.6f}, MSE = {mse_loss.item():.6f}, Physics = {phys_loss.item():.6f}")

# # After training
# print("Final logged losses:")
# print(loss_log)


model = PINN(input_dim=X.shape[1])
optimizer = optim.Adam(model.parameters(), lr=1e-3)
epochs = 1000
lambda_phys = 0.8 
loss_log = {
    "total": [],
    "mse": [],
    "phys": []
}

for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    
    y_pred = model(X)
    y_pred_norm = y_pred / torch.sum(y_pred)

    mse_loss = nn.MSELoss()(y_pred_norm, y_true)

    energies = ising_energy(X)
    phys_loss = physics_loss(y_pred_norm, energies, E0)

    scale_factor = mse_loss.item() / (phys_loss.item() + 1e-8)
    phys_loss_scaled = phys_loss * scale_factor

    total_loss = lambda_phys * phys_loss_scaled + (1 - lambda_phys) * mse_loss

    loss_log["total"].append(total_loss.item())
    loss_log["mse"].append(mse_loss.item())
    loss_log["phys"].append(phys_loss.item())  # log unscaled for analysis

    total_loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print(f"Epoch {epoch}: Total Loss = {total_loss.item():.6f}, MSE = {mse_loss.item():.6f}, Physics = {phys_loss.item():.6f} (Scale={scale_factor:.4e})")

# After training
print("Final logged losses:")
print(loss_log)
```

```python
import matplotlib.pyplot as plt

model.eval()
with torch.no_grad():
    y_pred = model(X)
    y_pred_norm = y_pred / torch.sum(y_pred)

y_pred_np = y_pred_norm.squeeze().numpy()
y_true_np = y_true.squeeze().numpy()

print(y_pred_np)
print(y_true_np)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(y_true_np, label="True $\psi^2$", marker='o', linestyle='none', alpha=0.6)
plt.plot(y_pred_np, label="Predicted $\psi^2$", marker='x', linestyle='none', alpha=0.6)
plt.xlabel("Configuration Index")
plt.ylabel("Probability")
plt.title("True vs Predicted Probability Distribution")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import r2_score
import torch.nn.functional as F

model.eval()
with torch.no_grad():
    y_pred = model(X)
    y_pred_norm = y_pred / torch.sum(y_pred)

    y_true_np = y_true.numpy().flatten()
    y_pred_np = y_pred_norm.numpy().flatten()

kl_div = F.kl_div(torch.log(y_pred_norm + 1e-8), y_true, reduction='batchmean')  # Add epsilon for safety
print(f"KL Divergence: {kl_div.item():.6f}")

r2 = r2_score(y_true_np, y_pred_np)
print(f"R Score: {r2:.6f}")
```
